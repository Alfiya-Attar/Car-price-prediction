### DATA PRE PROCESSING

```python
# Importing libraries:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading Dataset:
data=pd.read_csv("car_price_prediction.csv")
data
```

```python
# To see first five rows:
data.head()
```

```python
# To check shape of data: 
data.shape
```

```python
# To get data information:
data.info()
```

```python
# To check null values:
data.isnull().sum()
```

```python
#changing representation of doors column :
data['Doors'] = data['Doors'].str.replace('04-May','4-5')
data['Doors'] = data['Doors'].str.replace('02-Mar','2-3')
print(data.Doors)
```

```python
# Replace dash with zero:
data["Levy"].replace("-",0)
```

```python
#Removing km in Mileage column an coverting it to integer: 
data['Mileage'] = data['Mileage'].str.replace('km', '')
data['Mileage'] = data['Mileage'].astype(str).astype(int)
```

```python
data
```

```python
#Removing Turbo in engine volume column an coverting it to integer:
data['Engine volume'] = data['Engine volume'].str.replace('Turbo', '')
data['Engine volume'] = data['Engine volume'].astype(str).astype(float)
```

```python
data
```

```python
# Removing columns which we don't need:
data.drop(['ID','Prod. year'],axis = 1, inplace = True)
data
```

### MACHINE LEARNING MODEL

```python
#libraries for preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
#libraries for evaluation
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
#Encoding data:
en=LabelEncoder()
for i in data.columns:
    data[i]=en.fit_transform(data[i])
```

```python
data
```

```python
#scaling:
col = ['Price','Levy','Engine volume']
feature = data[col]
scaler = StandardScaler().fit(feature.values)
feature = scaler.transform(feature.values)
data[col] = feature
```

```python
data.head()
```

```python
#splitting data into x:
x=data.iloc[:,1:16]
x
```

```python
#splitting data into y:
y=data.iloc[:,0]
y
```

```python
# XGboost model:
from xgboost import XGBRegressor
import sklearn.metrics as metrics
```

```python
#Splitting data into train and test:
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30)
```

```python
#Fitting xgboost model:
xgb_model = XGBRegressor().fit(xtrain, ytrain)
```

```python
xgb_pred = xgb_model.predict(xtest)
xgb_pred
```

#### Evaluating model using different metrics

```python
print("FOR XGBoost model:\n")
print('Mean Absolute Error:', round(metrics.mean_absolute_error(ytest,xgb_pred ),2))
print('Mean Squared Error:', round(metrics.mean_squared_error(ytest,xgb_pred ),2))
print('Root Mean Squared Error:', round(np.sqrt(metrics.mean_squared_error(ytest,xgb_pred )),2))
print('R-square:',r2_score(ytest,xgb_pred ))
```

```python
#training accuracy
xgb_model.score(xtrain,ytrain)
```

```python
#testing accuracy:
xgb_model.score(xtest,ytest)
```

**CONCLUSION: XGBoost Model is giving 70.40% accuracy.**

```python
# K NeighborS Classifier:
from sklearn.neighbors import KNeighborsRegressor
```

```python
KNN=KNeighborsRegressor(n_neighbors=5) 
```

```python
#Fitting K Neighbors classifier:
knn_model=KNN.fit(xtrain,ytrain)
```

```python
knn_pred=KNN.predict(xtest)
knn_pred
```

#### Evaluating model using different metrics

```python
print("K Neighbors Classifier:\n")
print('Mean Absolute Error:', round(metrics.mean_absolute_error(ytest,knn_pred ),2))
print('Mean Squared Error:', round(metrics.mean_squared_error(ytest,knn_pred ),2))
print('Root Mean Squared Error:', round(np.sqrt(metrics.mean_squared_error(ytest,knn_pred )),2))
print('R-square:',r2_score(ytest,knn_pred ))
```

```python
#Training accuracy:
knn_model.score(xtrain,ytrain)
```

```python
#Testing accuracy:
knn_model.score(xtest,ytest)
```

**CONCLUSION: KNeighbors Model is giving 21.92% accuracy.**

**CONCLUSION: SO WE CAN SAY THAT XGBoost IS BEST MODEL FOR CARS PRICE PREDICTION.**
